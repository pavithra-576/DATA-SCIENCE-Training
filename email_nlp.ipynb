```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to address LookupError
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
data['spam'].count()
```

```python
v=set(stopwords.words('english'))
import re
bag=[]
lm=PorterStemmer()
for i in range(len(data)):
  clean_text=re.sub('[^a-zA-Z]',' ',data['text'][i])
  clean_text=clean_text.lower()
  words=clean_text.split()
  processed_words=[lm.stem(word) for word in words if not word in v]
  processed_text=' '.join(processed_words)
  bag.append(processed_text)
```

```python
text = data['text'][0] # Example: Assigning the first email for processing
text=word_tokenize(text)
text=[word for word in text if not word in v]
```

```python
print(bag)
```

```python
h=len(bag)
print(h)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5728)
X=cv.fit_transform(bag).toarray()
y=data['spam']
```

```python
print(type(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
ytest=y_test.values
ytest
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
